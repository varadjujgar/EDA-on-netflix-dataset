### Subtask:
Process the 'duration' column to extract numerical values for movies (minutes) and TV shows (seasons), and then visualize their distributions.


**Reasoning**:
I will create two new columns, 'duration_minutes' and 'duration_seasons', and populate them by extracting numerical values from the 'duration' column based on the 'type' of content (Movie or TV Show). This involves using regular expressions to parse the string durations and converting them to appropriate numeric types, as specified in the instructions.



````python
df['duration_minutes'] = df.apply(lambda row: int(row['duration'].split(' ')[0]) if 'min' in row['duration'] else None, axis=1)
df['duration_seasons'] = df.apply(lambda row: int(row['duration'].split(' ')[0]) if 'Season' in row['duration'] else None, axis=1)

# Convert to numeric, coercing errors
df['duration_minutes'] = pd.to_numeric(df['duration_minutes'], errors='coerce')
df['duration_seasons'] = pd.to_numeric(df['duration_seasons'], errors='coerce')

print("New 'duration_minutes' and 'duration_seasons' columns created and populated.")
print(df[['type', 'duration', 'duration_minutes', 'duration_seasons']].head())
````

**Reasoning**:
I will create a histogram for the 'duration_minutes' column to visualize the distribution of movie durations, and another histogram for the 'duration_seasons' column to visualize the distribution of TV show seasons, as specified by the instructions.



````python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['duration_minutes'].dropna(), bins=30, kde=True)
plt.title('Distribution of Movie Durations (Minutes)')
plt.xlabel('Duration (Minutes)')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df['duration_seasons'].dropna(), bins=df['duration_seasons'].max(), kde=False)
plt.title('Distribution of TV Show Seasons')
plt.xlabel('Number of Seasons')
plt.ylabel('Count')
plt.xticks(range(1, int(df['duration_seasons'].max()) + 1))
plt.show()
````

### Comparison of Content Type Trends: Release Year vs. Added Year

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Ensure 'date_added' is in datetime format and 'add_year' is extracted
# This might have been done already, but it's good practice to ensure for this specific comparison.
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['add_year'] = df['date_added'].dt.year

# Data preparation for Release Year trends
content_yearly_release = df.groupby(['release_year', 'type']).size().reset_index(name='count')

# Data preparation for Added Year trends
content_yearly_added = df.groupby(['add_year', 'type']).size().reset_index(name='count')

# Plot 1: Content Type Trends by Release Year
plt.figure(figsize=(14, 7))
sns.lineplot(data=content_yearly_release, x='release_year', y='count', hue='type')
plt.title('Content Type Trends Over Release Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.legend(title='Content Type')
plt.grid(True)
plt.show()

# Plot 2: Content Type Trends by Added Year
plt.figure(figsize=(14, 7))
sns.lineplot(data=content_yearly_added, x='add_year', y='count', hue='type')
plt.title('Content Type Trends Over Years Added to Netflix')
plt.xlabel('Year Added')
plt.ylabel('Number of Titles')
plt.legend(title='Content Type')
plt.grid(True)
plt.show()
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Ensure 'df' is defined by reloading the dataset
df = pd.read_csv('netflix_titles_nov_2019.csv')

# Ensure 'date_added' is in datetime format and 'add_year' is extracted
# This might have been done already, but it's good practice to ensure for this specific comparison.
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['add_year'] = df['date_added'].dt.year

# Data preparation for Release Year trends
content_yearly_release = df.groupby(['release_year', 'type']).size().reset_index(name='count')

# Data preparation for Added Year trends
content_yearly_added = df.groupby(['add_year', 'type']).size().reset_index(name='count')

# Plot 1: Content Type Trends by Release Year
plt.figure(figsize=(14, 7))
sns.lineplot(data=content_yearly_release, x='release_year', y='count', hue='type')
plt.title('Content Type Trends Over Release Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.legend(title='Content Type')
plt.grid(True)
plt.show()

# Plot 2: Content Type Trends by Added Year
plt.figure(figsize=(14, 7))
sns.lineplot(data=content_yearly_added, x='add_year', y='count', hue='type')
plt.title('Content Type Trends Over Years Added to Netflix')
plt.xlabel('Year Added')
plt.ylabel('Number of Titles')
plt.legend(title='Content Type')
plt.grid(True)
plt.show()
````

### Observations and Comparison:

**1. Overall Trend:**
*   **Release Year Plot:** Shows a historical trend of content production. We can see a gradual increase in both movies and TV shows released over the decades, with a significant surge in production in the 2000s, especially peaking around 2016-2019. This reflects a general increase in global content production.
*   **Added Year Plot:** Reflects Netflix's content acquisition strategy. It shows a steep and continuous increase in the number of titles added to Netflix, particularly from around 2015 onwards, indicating a period of aggressive content expansion by the platform.

**2. Dominance of Movies:**
*   In both plots, **Movies consistently outnumber TV Shows**. This indicates that Netflix's library, based on this dataset, is predominantly movie-centric, both in terms of content released historically and content added to the platform.

**3. Gap Between Release and Added Years:**
*   There is a clear distinction where content, especially older movies, was released many years before being added to Netflix. For instance, the 'Release Year' plot shows content from as early as the 1920s, whereas the 'Added Year' plot starts showing significant additions much later, primarily from the mid-2000s.
*   For newer content (especially from the mid-2010s onwards), the gap between release year and added year becomes much smaller, suggesting Netflix is acquiring newer content closer to its release date, or even producing its own content.

**4. Growth Trajectory:**
*   The **'Added Year' plot shows a more dramatic and recent exponential growth** in content volume compared to the 'Release Year' plot, especially for the years leading up to 2019. This highlights Netflix's rapid expansion in content offering.
*   The 'Release Year' plot's increase is more spread out over a longer historical period.

**5. Implications:**
*   The 'Release Year' trend gives insight into the historical output of the entertainment industry.
*   The 'Added Year' trend is more indicative of Netflix's business growth, content acquisition strategies, and the expansion of its platform over time. The sharp rise in added content, particularly post-2015, points to a period of intense competition in the streaming market and Netflix's response to it.
